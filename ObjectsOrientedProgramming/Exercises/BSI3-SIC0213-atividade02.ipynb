## Atividade Prática 2: Palavra-chave self

1. Qual palavra-chave (keyword) você usaria para ter acesso às propriedades e métodos de uma classe estando dentro dela?

  c) A palavra-chave self

```python
# Adicionar ao método Hello a capacidade de acessar as propriedades
# dentro da classe

class Usuario():
  nome = ""
  sobrenome = ""

  # Incluir o primeiro nome na mensagem do método
  def Hello (self):
    print (f"Hello, {self.nome}")

# Adicionar um objeto da classe usuário que receba um nome e sobrenome,
# e usar o método Hello da classe
usuario = Usuario()
usuario.nome = "Jonnie"
usuario.sobreonome = "Bravo"
usuario.Hello()
```
