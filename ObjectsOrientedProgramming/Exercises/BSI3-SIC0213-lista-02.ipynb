### Classe Pessoa 01

Crie uma classe Pessoa com os atributos nome e idade. Crie uma classe Aluno que herda de Pessoa e adicione o atributo nota. Crie um método para imprimir as informações da Pessoa e um método para imprimir as informações do Aluno.

```python
class Pessoa():
  def __init__ (self, nome, idade):
    self._nome = nome #string
    self._idade = idade #int

  def __str__ (self):
    return f"\nNome: {self._nome}\nIdade: {self._idade}"

class Aluno (Pessoa):
  
  def __init__(self, nome, idade, nota):
    super().__init__(nome, idade)
    self.__nota = nota
  
  def mudarNota (self, nota):
    self.__nota = nota

  def __str__ (self):
    return f"\nNome: {self._nome}\nIdade: {self._idade}\nNota: {self.__nota}"
```

### Classe Veículo

Crie uma classe Veiculo com os atributos marca, modelo e ano. Crie classes filhas Carro e Moto que adicionam o atributo quantidade_de_portas e cilindradas, respectivamente. Crie um método para imprimir as informações do Veiculo e um método para imprimir as informações do Carro e da Moto.

```python
class Veiculo():
    def __init__ (self, marca, modelo, ano):
      self._marca = marca
      self._modelo = modelo
      self._ano = ano
    
    def __str__ (self):
      return f"Marca: {self._marca}\nModelo: {self._modelo}\nAno: {self._ano}\n" 

class Carro (Veiculo):
  def __init__ (self, marca, modelo, ano, portas):
    super().__init__(marca, modelo, ano)
    self.__portas = portas

  def __str__ (self):
    return f"{super().__str__()}Portas: {self.__portas}\n"

class Moto (Veiculo):
  def __init__ (self, marca, modelo, ano, cilindradas):
    super().__init__(marca, modelo, ano)
    self.__cilindradas = cilindradas

  def __str__ (self):
    return f"{super().__str__()}Cilindradas: {self.__cilindradas}\n"
```

```python
# TESTES

veiculinho = Veiculo('VolksWagen','Gol','2010')
print(veiculinho)
carrinho = Carro('Fiat','Uno','2017','4')
print(carrinho)
motinho = Moto ('Honda','Biz','2012','125')
print(motinho)
```

### Classe Empregado
Implemente o nDiagrama de Classes de Empregado ilustrado abaixo. Observações:
 - **O aumento do empregado é:** salário * percentual
 - **O aumento do chefe é:** salario * percentual + beneficio
 - **O aumento do estagiário é:** salario * percentual - descontos


image.png


```python
class Empregado():
  def __init__ (self, codigo, nome, email, salario):
    self.__codigo = codigo #int
    self.__nome = nome #string
    self.__email = email #string
    self.__salario = salario #float
  
  def getSalario (self):
    return self.__salario
  
  def aumentoSalario (self, percentual):
    self.__salario = self.__salario + ((self.__salario * percentual)/100)
  
class Chefe(Empregado):
  __beneficio = 0 #float

  def aumentoSalario (self, beneficio, percentual):
    salario = self.getSalario()
    self.__beneficio = beneficio
    self.__salario = salario + self.__beneficio + ((salario * percentual)/100)
    return self.__salario

class Estagiario(Empregado):
  __descontos = 0 #int

  def aumentoSalario (self, desconto, percentual):
    salario = self.getSalario()
    self.__descontos = percentual
    self.__descontos = desconto
    self.__salario = salario - self.__descontos + ((salario * percentual)/100)
    return self.__salario
```

```python
# TESTES
empregadinho = Empregado (92187, 'Maria', 'maria@gmail.com', 1350.0)
empregadinho.aumentoSalario(10)
print(empregadinho.getSalario())

chefinho = Chefe (92382, 'Lucas','lucas@gmail.com', 1500.0)
print(chefinho.aumentoSalario(50,7))

estagiariozinho = Estagiario(93979, 'Ariane', 'ariane@gmail.com', 500.0)
print(estagiariozinho.aumentoSalario(15, 9))
```

```python
class Ingresso ():
  def __init__ (self, valor):
    self._valor = valor
  
  def imprimeValor(self):
    print(f"\033[31mValor:\033[0m {self._valor}")

class Normal (Ingresso):
  def imprimeIngressoNormal(self):
    print (f"\033[31mValor Normal:\033[0m R${self._valor}")

class Vip (Ingresso):
  def __init__ (self, valor, adicional):
    super().__init__(valor)
    self._adicional = adicional
  
  def imprimeValor (self):
    print (f"\033[31mValor VIP:\033[0m R${self._valor + self._adicional}")

class CamaroteInferior (Vip):
  def __init__ (self, valor, adicional, local):
    super().__init__(valor, adicional)
    self.__local = local
  
  def imprimeLocal (self):
    print (f"\033[31mLocal do Camarote:\033[0m {self.__local}")

class CamaroteSuperior (Vip):
  def __init__ (self, valor, adicional, plus):
    super().__init__(valor, adicional)
    self.__adicionalPlus = plus
  
  def imprimeValor (self):
    print (f"\033[31mValor do Camarote:\033[0m R${self._valor + self._adicional + self.__adicionalPlus}")
```

```python
class Pessoa ():
  def __init__ (self, nome, idade):
    self._nome = nome
    self._idade = idade
  
class Rica (Pessoa):
  def __init__ (self, nome, idade, dinheiro):
    super().__init__ (nome, idade)
    self.__dinheiro = dinheiro
  
  def fazCompras (self, compra):
    self.__dinheiro = self.__dinheiro - compra
    return f"Compras feitas! Riqueza = R${self.__dinheiro}"

class Pobre (Pessoa):
  def trabalha (self):
    return f"{self._nome} está trabalhando..."

class Miseravel (Pessoa):
  def mendiga (self):
    return f"{self._nome} está mendigando..."
```

```python
# TESTES
pessoinha = Pessoa ("Maria", 21)
print(pessoinha._nome)
print(pessoinha._idade)

riquinho = Rica ("Chico", 20, 10000)
print(riquinho.fazCompras(100))
print(riquinho.fazCompras(2300))

pobrinho = Pobre ("Maria",21)
print(pobrinho.trabalha())

miseravelzinho = Miseravel ("Ze", 30)
print(miseravelzinho.mendiga())
```
