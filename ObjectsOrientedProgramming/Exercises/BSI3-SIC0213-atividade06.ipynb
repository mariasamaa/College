5. Agora, vamos acrescentar à classe Admin algum código. Primeiro, adicione um método público chamado de escrevaNome() e faça com que ele retorne a string: "Admin".
6. Adicione à classe Admin outro método público, digaOla(), que retorne a string "Olá Admin,

7. Crie um objeto chamado admin1 fora da classe Admin. Defina seu nome para "Baltazar" e diga olá ao usuário (...)

```python
class Usuario(): #classe mae

  __nomeUsuario = ""

  def setDefinirNome (self, nome):
    __nomeUsuario = nome
  
class Admin(Usuario):

  def escreverNome (self): #classe filha
    return "Admin"
  
  def digaOla(self):
    return f"Olá Admin, {self.__nomeUsuario}"

admin1 = Admin()
print (admin1.escreverNome())
admin1.setDefinirNome('Baltazar')
admin1.digaOla()
```

7. (...) Você vê algum problema?

8. Qual é a causa do problema?
  - O atributo nome de usuário foi definido com privado, por isso a classe não pode acessar;
9. Altere o código para corrigir o problema.
10. Tente escrever a solução com um método getter dentro da classe mãe que pode ser usado a partir da classe filha.

```python
class Usuario(): #classe mae

  _nomeUsuario = ""

  def getDefinirNome (self):
    return self._nomeUsuario
  
  def setDefinirNome(self,nome):
    self._nomeUsuario = nome
  
class Admin(Usuario):

  def escreverNome (self):
    return "Admin"
  
  def digaOla(self):
    return f"Olá Admin, {self.getDefinirNome()}"

admin1 = Admin()
admin1.setDefinirNome("Baltazar")
admin1.digaOla()
```
