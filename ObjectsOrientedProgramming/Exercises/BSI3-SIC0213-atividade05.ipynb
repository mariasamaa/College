## Atividade Prática 5: Métodos Mágicos

```python
# criar a classe Usuario
class Usuario():

  # incluir método construtor para definir valores das propriedades primeiroNome
  # e ultimoNome
  def __init__ (self, primeiroNome, ultimoNome):
    self.__primeiroNome = primeiroNome
    self.__ultimoNome = ultimoNome

  # incluir um método público que retorna o nome completo do usuário
  def getNomeCompleto(self):
    return f"Seu nome completo é {self.__primeiroNome} {self.__ultimoNome}"
  
# criar um novo objeto e passar ao contrutor os valores de nome
usuario1 = Usuario("Jonny","Bravo")
print (usuario1.getNomeCompleto())
```
