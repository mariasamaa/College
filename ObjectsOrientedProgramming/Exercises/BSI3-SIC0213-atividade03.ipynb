```python
class Usuario():

  __primeiroNome = ""
  
  def setPrimeiroNome (self, nome):
    self.__primeiroNome = nome

  def getPrimeiroNome (self):
    return f"Seu nome é {self.__primeiroNome}"
  
usuario1 = Usuario()
usuario1.setPrimeiroNome("Joe")
print (usuario1.getPrimeiroNome())
```

```python
class Empregado():
  __nome = ""
  __salario = ""
  projeto = ""

  def setTrabalho (self, nome, salario):
    self.__nome = nome
    self.__salario = salario

  def getTrabalho (self):
    return f"{self.__nome} está trabalho no projeto {self.projeto}"

  def getMostrar (self):
    return f"Nome: {self.__nome} // Salário: R${self.__salario}"

empregado1 = Empregado()
empregado1.setTrabalho("Maria","1200")
empregado1.projeto = "atividade de P.O.O."

print (empregado1.getMostrar())
print (empregado1.getTrabalho())
```

```python
class Robo():
  __nome = ""
  __ano_construcao = ""

  def setDiga_alo (self, nome, ano):
    self.__nome = nome
    self.__ano_construcao = ano
  
  def getDiga_alo (self):
    return f"""O nome do robo cadastrado é {self.__nome}.
Ele foi constrído em {self.__ano_construcao}."""

robo1 = Robo()
robo1.setDiga_alo("BMO","2007")
print (robo1.getDiga_alo())
```

```python
class Laptop():
  __preco = "1200"

  def getPreco (self):
    return f"Preço do Laptop = R${self.__preco}"
  
  def setPreco (self, preco):
    self.__preco = preco
  
laptop1 = Laptop()
print (laptop1.getPreco())

laptop2 = Laptop()
laptop2.setPreco ("3999")
print (laptop2.getPreco())
```
