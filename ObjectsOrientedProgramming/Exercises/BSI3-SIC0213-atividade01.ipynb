4. Qual das alternativas abaixo melhor explica o termo "método"?

   a) Uma função dentro de uma classe.

```python
# Criar a classe Usuario, adicionar os métodos de primeiro e último nome e um 
# método que retorne "Olá"

class Usuario():
  nome = ""
  sobrenome = ""

  def hello (self):
    print (f"Olá, {self.nome} {self.sobrenome}")

# Criar instância chamada "usuario1" e definir valores de primeiro/último nome

usuario1 = Usuario()
usuario1.nome = "Maria"
usuario1.sobrenome = "Caetano"
print(usuario1.nome, usuario1.sobrenome)

# Usar o método que retorna "Olá" com as variáveis de primeiro e último nome 
# para dizer "olá" ao usuário

usuario1.hello()

# Criar uma segunda instância com primeiro nome Jane e último nome Silva, 
# depois digá "olá" ao usuário

usuario2 = Usuario()
usuario2.nome = "Jane"
usuario2.sobrenome = "Silva"
usuario2.hello()
```
