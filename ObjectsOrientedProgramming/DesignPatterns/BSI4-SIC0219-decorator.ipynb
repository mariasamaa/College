# Atividade Prática 3: Decorator Pattern


**Leia atentamente o que está descrito abaixo.**

• Por outro lado, o sistema de pedidos do bar se tornou mais complexo e está difícil mantê-lo por conta da grande variedade de drinks;

• Em decorrência houve uma verdadeira explosão de classes para representar os diferentes tipos de drinks;

• Quero uma caipira de saquê, abacaxi, kiwi e açúcar;

• Quero uma caipira de vodka, morango e adoçante;

• Quero uma caipirinha (que é um drink pronto patenteado que não pode ser alterado – cachaça, limão, gelo e açúcar);

```python
from abc import ABC, abstractmethod
```

CLASSES BEBIDA

```python
class Bebidas(ABC): # classe componente abstrata
  def __init__ (self, custo=0, descricao=""):
      self._custo = custo
      self._descricao = descricao

  # método para custo
  @abstractmethod
  def getCusto(self):
    pass

  # método para descrição
  @abstractmethod
  def getDescricao(self):
    pass
```

```python
class Caipirinha(Bebidas): # classe componente concreta
  def __init__ (self):
    super().__init__(20.00, "\033[34mCAIPIRINHA\033[0m\nCachça\nLimão\nAçúcar\nGelo")

  # método para custo
  def getCusto(self):
    return self._custo # R$20,00

  # método abstrato para descrição
  def getDescricao(self):
    return self._descricao

  # capipirinha padrão: cachça, limão, açúcar, gelo
```

```python
class CaipiraSubway(Bebidas): # classe concreta componente
  def __init__ (self):
    super().__init__(0.00, "\033[34mCAIPIRINHA\033[0m")

  # método para custo
  def getCusto(self):
    return self._custo

  # método abstrato para descrição
  def getDescricao(self):
    return self._descricao
```

CLASSES PARA INGREDIENTRES DA CAIPIRINHA SUBWAY

```python
class Ingredientes(Bebidas, ABC): # classe decoradora base
  def __init__ (self, custo, descricao, bebida):
    super().__init__(custo, descricao)
    self._bebida = bebida

  # método para custo
  @abstractmethod
  def getCusto(self):
    pass

  # método abstrato para descrição
  @abstractmethod
  def getDescricao(self):
    pass
```

```python
class Vodca(Ingredientes): # decorador concreto
  def __init__ (self, bebida):
    super().__init__(20.0,"\nVodca", bebida)

  # método para custo
  def getCusto(self):
    return self._bebida.getCusto() + self._custo

  # método abstrato para descrição
  def getDescricao(self):
    return self._bebida.getDescricao() + self._descricao
```
