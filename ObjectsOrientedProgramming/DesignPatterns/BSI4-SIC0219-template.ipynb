### Atividade 01

Desenvolva o padrão “Template Method” criando uma classe abstrata que lê uma string, transforme-a e a imprima transformada. A transformação é delegada às subclasses. Implemente três subclasses, que são: uma que transforme a string toda para maiúsculo, outra que a transforme em tudo minúsculo e a última que a inverta.

---
Classe Abstrata

---

```python
class TransformandoString(ABC):
    def stringProcessada(self, stringzinha):
        stringTransformada = self.transformar(stringzinha)
        self.mostrarString(stringTransformada)

    @abstractmethod
    def transformar(self, stringzinha):
        pass

    def mostrarString(self, string_nova):
        print(string_nova)
```

---
Classs Concretas

---

```python
class TransformaMaisculo(TransformandoString):
    def transformar(self, stringzinha):
        return stringzinha.upper()

class TransformaMinusculo(TransformandoString):
    def transformar(self, stringzinha):
        return stringzinha.lower()

class TransformaReverso(TransformandoString):
    def transformar(self, stringzinha):
        return stringzinha[::-1]
```

---
Teste Cliente

---

```python
stringzinha = "TeMpLaTe MeThOd"

maiusculazinha = TransformaMaisculo()
minusculazinha = TransformaMinusculo()
reversazinha = TransformaReverso()

print("String Original:", stringzinha)

print(f"\nTransformando para Maiúscula:\n{maiusculazinha.transformar(stringzinha)}")

print(f"\nTransformando para Minúscula:\n{minusculazinha.transformar(stringzinha)}")

print(f"\nTransformando para Reversa:\n{reversazinha.transformar(stringzinha)}")
```

### Atividade 02

Suponha que queiramos comprar um telefone. Podemos comprá-lo de duas maneiras: podemos
comprá-lo online ou ir a uma loja de telefones e comprá-lo offline. As etapas envolvidas neste
processo completo são selecionar um telefone, depois disso seu telefone vai para a embalagem,
depois de embalado você fará o pagamento e por fim o telefone será entregue a você. Assim, você
pode ver que, quer compre um telefone online ou offline, as etapas envolvidas na compra são as
mesmas, embora sua implementação possa ser diferente. O ponto importante é que não podemos
alterar a ordem de execução porque não podemos comprar um telefone antes de selecioná-lo.
Portanto, neste caso, podemos criar um método de modelo (i.e. “Template Method”) que usará
métodos diferentes para comprar o telefone. Aqui podemos fornecer a implementação básica para as
mesmas etapas da compra de um telefone. Se as subclasses quiserem substituir esse método, elas
podem. A figura abaixo mostra todo este processo.

---
Baseado na figura acima, implemente o código usando o padrão de projeto “Template
Method”.

---
Classe Abstrata

---

```python

class ComprarTelefone(ABC):
    def comprarTelefone(self):
        self.escolherTelefone()
        self.embalarTelefone()
        self.pagarTelefone()
        self.entregarTelefone()

    @abstractmethod
    def escolherTelefone(self):
        pass

    @abstractmethod
    def embalarTelefone(self):
        pass

    @abstractmethod
    def pagarTelefone(self):
        pass

    @abstractmethod
    def entregarTelefone(self):
        pass
```

---
Classes Concretas

---

```python
class ComprarOnline(ComprarTelefone):
    def escolherTelefone(self):
        sleep(1)
        print("Escolhendo telefone online...")
        sleep(1)
        print ("Escolha processada")

    def embalarTelefone(self):
        sleep(1)
        print("Emabalando telefone para entrega...")
        sleep(1)
        print("Produto embalado com sucesso!")

    def pagarTelefone(self):
        sleep(1)
        print("Realizando o pagamento...")
        sleep(1)
        print("Pagamento processado com sucesso")

    def entregarTelefone(self):
        sleep(1)
        print("Seu telefone saiu para a entrega.")

class ComprarOffline(ComprarTelefone):
    def escolherTelefone(self):
        sleep(1)
        print("Escolhendo telefone na loja...")
        sleep(1)
        print ("Escolha processada")

    def embalarTelefone(self):
        sleep(1)
        print("Colocando seu telefone em uma bolsinha...")
        sleep(1)
        print("Produto embalado com sucesso!")

    def pagarTelefone(self):
        sleep(1)
        print("Realizando o pagamento...")
        sleep(1)
        print("Pagamento processado com sucesso")

    def entregarTelefone(self):
        sleep(1)
        print("Retire seu telefone no balcão de retirada!")
```

---
Testes Cliente

---

```python
comprinha_online = ComprarOnline()
comprinha_offline = ComprarOffline()

print("\033[34mPROCESSO DE COMPRAS ONLINE\033[0m")
comprinha_online.comprarTelefone()

sleep(1)
print("\n")

print("\033[34mPROCESSO DE COMPRAS OFFLINE\033[0m")
comprinha_offline.comprarTelefone()
```
