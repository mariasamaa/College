```python
from abc import ABC, abstractmethod
```

## Atividade 01

---
Produto Abstrato

---

```python
class Furniture(ABC):

  @abstractmethod
  def furnitureFunction(self):
    pass

  @abstractmethod
  def showStyle(self):
    pass
```

---
Produtos Concretos

---

```python
class ClassicCabinet(Furniture):

  def furnitureFunction(self):
    return "Cabinet"

  def showStyle(self):
    return "Classic"
```

```python
class ClassicChair(Furniture):

  def furnitureFunction(self):
    return "Chair"

  def showStyle(self):
    return "Classic"
```

```python
class ClassicDinningTable(Furniture):

  def furnitureFunction(self):
    return "Dinning Table"

  def showStyle(self):
    return "Classic"
```

```python
class ContemporaryCabinet(Furniture):

  def furnitureFunction(self):
    return "Cabinet"
  def showStyle(self):
    return "Contemporary"
```

```python
class ContemporaryChair(Furniture):

  def furnitureFunction(self):
    return "Chair"

  def showStyle(self):
    return "Contemporary"
```

```python
class ContemporaryDinningTable(Furniture):

  def furnitureFunction(self):
    return "Dinning Table"

  def showStyle(self):
    return "Contemporary"
```

```python
class ScandinavianCabinet(Furniture):

  def furnitureFunction(self):
    return "Cabinet"

  def showStyle(self):
    return "Scandinavian"
```

```python
class ScandinavianChair(Furniture):

  def furnitureFunction(self):
    return "Chair"

  def showStyle(self):
    return "Scandinavian"
```

```python
class ScandinavianDinningTable(Furniture):

  def furnitureFunction(self):
    return "Dinning Table"

  def showStyle(self):
    return "Scandinavian"
```

---
Fábrica Abstrata

---

```python
class FurnitureFactory(ABC):


  @abstractmethod
  def createCabinet(self):
    pass

  @abstractmethod
  def createChair(self):
    pass

  @abstractmethod
  def createDinningTable(self):
    pass
```

---
Fábricas concretas

---

```python
class ScandinavianFactory(FurnitureFactory):

  def createCabinet(self):
    return ScandinavianCabinet()

  def createChair(self):
    return ScandinavianChair()


  def createDinningTable(self):
    return ScandinavianDinningTable()
```

```python
class ClassicFactory(FurnitureFactory):

  def createCabinet(self):
    return ClassicCabinet()

  def createChair(self):
    return ClassicChair()

  def createDinningTable(self):
    return ClassicDinningTable()
```

```python
class ContemporaryFactory(FurnitureFactory):

  def createCabinet(self):
    return ContemporaryCabinet()

  def createChair(self):
    return ContemporaryChair()

  def createDinningTable(self):
    return ContemporaryDinningTable()
```

---
Testes Cliente

---

```python
# Fábrica Contemporânea
fabricazinha_contemporanea = ContemporaryFactory()
cadeirinha = fabricazinha_contemporanea.createChair()
mesinha = fabricazinha_contemporanea.createDinningTable()
armarinho = fabricazinha_contemporanea.createCabinet()

print(cadeirinha.furnitureFunction())
print(cadeirinha.showStyle())
print(mesinha.furnitureFunction())
print(mesinha.showStyle())
print(armarinho.furnitureFunction())
print(armarinho.showStyle())
print()

# Fábrica Clássica
fabricazinha_classica = ClassicFactory()
cadeirinha = fabricazinha_classica.createChair()
mesinha = fabricazinha_classica.createDinningTable()
armarinho = fabricazinha_classica.createCabinet()
print(cadeirinha.furnitureFunction())
print(cadeirinha.showStyle())
print(mesinha.furnitureFunction())
print(mesinha.showStyle())
print(armarinho.furnitureFunction())
print(armarinho.showStyle())
print()

# Fábrica Escandinava
fabricazinha_escandinava = ContemporaryFactory()
cadeirinha = fabricazinha_escandinava.createChair()
mesinha = fabricazinha_escandinava.createDinningTable()
armarinho = fabricazinha_escandinava.createCabinet()
print(cadeirinha.furnitureFunction())
print(cadeirinha.showStyle())
print(mesinha.furnitureFunction())
print(mesinha.showStyle())
print(armarinho.furnitureFunction())
print(armarinho.showStyle())
print()
```

## Atividade 2

---
Produto Abstrato

---

```python
class MotorVehicle(ABC):

  @abstractmethod
  def vehicleInfo(self):
    pass
```

```python
class ElectricVehicle(ABC):

  @abstractmethod
  def vehicleInfo(self):
    pass
```

---
Produtos Concretos

---

```python
class FutureMotorcycle(MotorVehicle):

  def vehicleInfo(self):
    return "Motor"
```

```python
class NextGenMotorcycle(MotorVehicle):

  def vehicleInfo(self):
    return "Motor"
```

```python
class FutureElectricCar(ElectricVehicle):

  def vehicleInfo(self):
    return "Electric"
```

```python
class NextGenElectricCar(ElectricVehicle):

  def vehicleInfo(self):
    return "Electric"
```

---
Fábrica Abstrata

---

```python
class Corporation(ABC):

  @abstractmethod
  def createMotorVehicle(self):
    pass

  @abstractmethod
  def createElectricVehicle(self):
    pass
```

```python
class FutureVehicleCorporation(Corporation):

  def createMotorVehicle(self):
    return FutureMotorcycle()

  def createElectricVehicle(self):
    return FutureElectricCar()
```

```python
class NextGenCorporation(Corporation):

  def createMotorVehicle(self):
    return FutureMotorcycle()

  def createElectricVehicle(self):
    return FutureElectricCar()
```

---
Teste Cliente

---

```python
fabricazinhadofuturo = FutureVehicleCorporation()
motinha = fabricazinhadofuturo.createMotorVehicle()
carrinho = fabricazinhadofuturo.createElectricVehicle()
print(motinha.vehicleInfo())
print(carrinho.vehicleInfo())

print()

fabricazinhadaproximageracao = NextGenCorporation()
motinha = fabricazinhadaproximageracao.createMotorVehicle()
carrinho = fabricazinhadaproximageracao.createElectricVehicle()
print(motinha.vehicleInfo())
print(carrinho.vehicleInfo())
```
