## Atividade 01

```python
from abc import ABC, abstractmethod
```

---
Classes Produto

---

```python
class MotorVehicle(ABC): # criar interface produto
  def __init__ (self, marca=None, modelo=None, cor=None):
    self.marca = marca
    self.modelo = modelo
    self.cor = cor

  @abstractmethod
  def build (self, marca, modelo, cor):
    pass

  def __str__ (self):
    return f"MARCA: {self.marca}\nMODELO: {self.modelo}\nCOR: {self.cor}\n"
```

```python
class Moto(MotorVehicle): # criar classe produto concreto

  def __init__ (self, marca, modelo, cor):
    super().__init__(marca, modelo, cor)

  def build(self, marca, modelo, cor):
    print ("Contruindo carro...")
```

```python
class Carro(MotorVehicle): # criar classe produto concreto

  def __init__ (self, marca, modelo, cor):
    super().__init__(marca, modelo, cor)

  def build(self, marca, modelo, cor):
    print ("Contruindo moto...")
```

---
Classes Criador

---

```python
class MotorVehicleFactory (ABC): # criar classe abstrata Creator

  @abstractmethod
  def create(self):
    pass

  @abstractmethod
  def createMotorVehicle(self):
    pass
```

```python
class CarroFactory(MotorVehicleFactory): # criar classe criador concreto
  def create(self):
    pass

  def createMotorVehicle(self, marca, modelo, cor):
    return Carro (marca, modelo, cor)
```

```python
class MotoFactory(MotorVehicleFactory): # criar classe criador concreto

  def create(self):
    pass

  def createMotorVehicle(self, marca, modelo, cor):
    return Moto (marca, modelo, cor)
```

---
Testes Cliente

---

```python
fabricazinhacarro = CarroFactory()
carrinho = fabricazinhacarro.createMotorVehicle('fiat', 'uno', 'vermelho')
print(carrinho)

fabricazinhamoto = MotoFactory()
motinha = fabricazinhamoto.createMotorVehicle('honda','CBR','preta')
print(motinha)
```

```python
class Product(ABC): # criar interface geral do produto
  def __init__ (self, color=None, width=None, height=None, depth=None):
    self.color = color
    self.witdh = width
    self.height = height
    self.depth = depth

  @abstractmethod
  def getColor(self, color):
    pass

  @abstractmethod
  def getWidth(self, width):
    pass

  @abstractmethod
  def getHeight(self, height):
    pass

  @abstractmethod
  def getDepth(self,depth):
    pass

  def __str__ (self): #método "ToString"
    return f"COLOR: {self.color}\nWITDH: {self.witdh}cm\nHEIGHT: {self.height}cm\nDEPTH: {self.depth}cm\n"
```

```python
class Armchair(Product,ABC): # criar interface do produto
  def __init__ (self, color=None, width=None, height=None, depth=None):
    super().__init__(color,width,height,depth)

  @abstractmethod
  def getColor(self, color):
    pass

  @abstractmethod
  def getWidth(self, width):
    pass

  @abstractmethod
  def getHeight(self, height):
    pass

  @abstractmethod
  def getDepth(self,depth):
    pass
```

```python
class CoffeeTable(Product,ABC): # criar interface do produto
  def __init__ (self, color=None, width=None, height=None, depth=None):
    super().__init__(color,width,height,depth)

  @abstractmethod
  def getColor(self, color):
    pass

  @abstractmethod
  def getWidth(self, width):
    pass

  @abstractmethod
  def getHeight(self, height):
    pass

  @abstractmethod
  def getDepth(self,depth):
    pass
```

```python
class Sofa(Product,ABC): # criar interface do produto
  def __init__ (self, color=None, width=None, height=None, depth=None):
    super().__init__(color,width,height,depth)

  @abstractmethod
  def getColor(self, color):
    pass

  @abstractmethod
  def getWidth(self, width):
    pass

  @abstractmethod
  def getHeight(self, height):
    pass

  @abstractmethod
  def getDepth(self,depth):
    pass
```

Classes Creator

---

```python
class FurnitureFactory(ABC): # criar classe abstrata Creator

  @abstractmethod
  def makeArmchair(self):
    pass

  @abstractmethod
  def makeCoffeeTable(self):
    pass

  @abstractmethod
  def makeSofa(self):
    pass
```

```python
class RetroFurnitureFactory(FurnitureFactory): # criar classe concreta Creator

  def makeArmchair(self, color, width, height, depth):
    return RetroArmchair(color, width, height, depth)

  def makeCoffeeTable(self, color, width, height, depth):
    return RetroCoffeeTable(color, width, height, depth)

  def makeSofa(self, color, width, height, depth):
    return RetroSofa(color, width, height, depth)
```

```python
class ModernFurnitureFactory(FurnitureFactory): # criar classe concreta Creator

  def makeArmchair(self, color, width, height, depth):
    return ModernArmchair(color, width, height, depth)

  def makeCoffeeTable(self, color, width, height, depth):
    return ModernCoffeeTable(color, width, height, depth)

  def makeSofa(self, color, width, height, depth):
    return ModernSofa(color, width, height, depth)
```
